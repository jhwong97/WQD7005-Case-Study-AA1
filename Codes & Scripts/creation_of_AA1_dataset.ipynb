```python
# Import the necessaru python package
import pandas as pd
import random
```

```python
# Read the original_data from Kaggle for 2023 portion
filepath = r"./original_data_2023.csv"

# Drop any rows that contain NA value
df = pd.read_csv(filepath, sep=',').dropna()
df.head(10)
```

```python
# Create an empty data frame with below column names
schema = {"CustomerID" : [],
          "Age" : [],
          "Gender" : [],
          "Location" : [],
          "Occupation" : [],
          "MembershipLevel" : [],
          "TotalPurchases" : [],
          "TotalSpent" : [],
          "FavouriteCategory" : [],
          "LastPurchaseDate" : [],
          "PaymentMethod" : [],
          "Churn" : []}
final_df = pd.DataFrame(schema)
final_df
```

```python
# Assuming 'df' is your dataframe and 'customer_name' is the column containing customer names
unique_customer_names = list(df['Customer'].unique())

# To extract the customer name from the original dataset and extract the necessary information to create AA1 dataset
for name in unique_customer_names:
    df_test = df[df['Customer']==name].copy()
    
    # Information on Purchase Date
    df_test['PurchaseDate'] = pd.to_datetime(df_test['PurchaseDate'], errors='coerce')
    df_test['PurchaseDate'] = df_test['PurchaseDate'].dt.date
    
    # Information on customer ID
    customer_id = df_test['CustomerID'].max().astype(int)
    # Information on customer Age
    customer_age = df_test['Age'].max().astype(int)
    # Information on customer gender
    gender_list = ['Male', 'Female']
    customer_gender = random.choice(gender_list)
    
    # Information on customer purchasing location
    location_list = ['Johor', 'Kedah', 'Kelantan', 'Malacca', 'Negeri Sembilan', 
                     'Pahang', 'Penang', 'Perak', 'Perlis', 'Sabah', 'Sarawak',
                     'Selangor', 'Terengganu', 'Kuala Lumpur', 'Putrajaya']
    customer_location = random.choice(location_list)
    
    # Information on customer e-commerce membership level
    membership_level_list = ['Bronze', 'Silver', 'Gold', 'Platinum']
    membership_level = random.choice(membership_level_list)
    
    # Information on customer total purchases on e-commerce
    total_purchases = df_test['TotalPurchases'].sum().astype(int)
    
    # Information on customer favourite category
    category_purchases = df_test.groupby('ProductCategory')['TotalPurchases'].sum()
    favourite_category = category_purchases.idxmax()
    
    # Information on customer total spent
    total_spent = df_test['TotalSpent'].sum()
    
    # Information on customer favourite payment method
    payment_method_list = ['Cash', 'Credit Card', 'TnG']
    payment_method = random.choice(payment_method_list)
    
    # Information on customer occupation
    common_occupations = [
        "Information Technology",
        "Education",
        "Healthcare",
        "Engineering",
        "Finance",
        "Medicine",
        "Marketing",
        "Design",
        "Customer Service",
        "Arts",
        "Writing",
        "Law Enforcement",
        "Architecture",
        "Business Analysis",
        "Human Resources",
        "Pharmacy",
        "Piloting",
        "Research",
        "Entrepreneurship",
        "Social Work",
        "Event Planning",
        "Real Estate",
        "Fitness",
        "Photography",
        "Plumbing",
        "Construction",
        "Dentistry",
        "Journalism",
        "Student"
    ]
    customer_occupation = random.choice(common_occupations)
    
    # Information on customer last purchase date
    last_purchase = df_test['PurchaseDate'].max()

    # Information on customer churn
    churn = df_test['Churn'].max().astype(int)
    
    # Create a dataframe encompasses all the previous stated information
    customer_details = pd.DataFrame({"CustomerID" : [customer_id],
                        "Age" : [customer_age],
                        "Gender" : [customer_gender],
                        "Location" : [customer_location],
                        "Occupation" : [customer_occupation],
                        "MembershipLevel" : [membership_level],
                        "TotalPurchases" : [total_purchases],
                        "TotalSpent" : [total_spent],
                        "FavouriteCategory" : [favourite_category],
                        "LastPurchaseDate" : [last_purchase],
                        "PaymentMethod" : [payment_method],
                        "Churn" : [churn]})
    
    # Concat the dataframe into the empty dataframe
    final_df = pd.concat([final_df, customer_details], ignore_index=True)
    final_df['CustomerID'] = final_df['CustomerID'].astype(int)
    final_df['TotalPurchases'] = final_df['TotalPurchases'].astype(int)
    final_df['Age'] = final_df['Age'].astype(int)
    final_df['Churn'] = final_df['Churn'].astype(int)
    final_df['TotalSpent'] = final_df['TotalSpent'].astype(int)
```
